```python
import pandas as pd


def find_columns_with_outliers(df):
    outlier_columns = []
    
    for column in df.select_dtypes(include=['number']).columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        
        lower_limit = q1 - 1.5 * iqr
        upper_limit = q3 + 1.5 * iqr
        
        # Identify outliers in the column
        outliers = df[(df[column] < lower_limit) | (df[column] > upper_limit)]
        
        if len(outliers) > 0:
            outlier_columns.append(column)
            print(f'Column "{column}" has outliers:')
            print(f'  Lower limit: {lower_limit}')
            print(f'  Upper limit: {upper_limit}')
            print(f'  Number of outliers: {len(outliers)}\n')
    
    return outlier_columns

# Find columns with outliers
columns_with_outliers = find_columns_with_outliers(df)
print('Columns with outliers:', columns_with_outliers)
```

```python
import pandas as pd

def remove_outliers_for_adults(df):
    column = 'total_of_special_requests'
    
    # Calculate Q1, Q3, and IQR
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    
    # Calculate lower and upper limits
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    
    print(f'Column "{column}" has outliers:')
    print(f'  Lower limit: {lower_limit}')
    print(f'  Upper limit: {upper_limit}')
    
    # Filter out the outliers
    df_cleaned = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]
    
    # Print the number of outliers removed
    print(f'Number of outliers removed: {df.shape[0] - df_cleaned.shape[0]}')
    
    return df_cleaned

# Call the function and clean the DataFrame
df_cleaned = remove_outliers_for_adults(df)

# Output the shape of the cleaned DataFrame
print('Shape of DataFrame after removing outliers:', df_cleaned.shape)
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not cancelled','cancelled'], df['is_canceled'].value_counts(), edgecolor='k')
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df,palette = 'Blues')
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#From above in city hotel maintainance and other facilities can be reason of cancellation
#And in resort hotel adr column i.e 'Average Daily rate' can be reason
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#In city hotel 41% got canceled
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#Blue line represents the resort hotel and orange line for city hotel
#Spikes reprents weekends or holidays that means on weekends prices got higher
```
